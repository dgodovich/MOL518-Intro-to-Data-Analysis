```python
chicken_f = open('Chicken-Weights.csv') # open the file
```

```python
chicken_flist=[];
for line in chicken_f:
    chicken_ff = line.rstrip()
#    print(ff)
    chicken_flist.append(chicken_ff.rsplit(','))

chicken_flist.pop(0)
```

```python
feeds = {}
for line in chicken_flist:
    if line[0] in feeds:
        feeds[line[0]].append(float(line[1]))
    else:
        feeds[line[0]] = []
        feeds[line[0]].append(float(line[1]))

# print(feeds)
```

```python
# plot the datasets
meanlist = []
stdlist = []
varlist = []
for key in feeds:
    meanlist.append(np.mean(feeds[key])) # compute the mean
    stdlist.append(np.std(feeds[key])) # compute the standard deviation
    varlist.append(np.var(feeds[key])) # compute the variance

print(meanlist)
print(varlist)
```

```python
fig, ax = plt.subplots()
ebplot=plt.errorbar([1,2,3,4,5,6],meanlist,stdlist,xerr=None,fmt='o')
plt.xlabel('Type of Feed')
plt.ylabel('Weight (g)')
ax.set_xticks([1,2,3,4,5,6], feeds.keys())
```

```python
ebplot.get_children()
```

```python
# calculate a matrix of p-values
pvalmx = []
counter = 0
for key in feeds:
    pvalmx.append([])
    for key2 in feeds:
        pval = sp.stats.ttest_ind(feeds[key], feeds[key2], axis=0, equal_var=False) # use Welch's t test since the variances are not equal
        pvalmx[counter].append(pval.pvalue)
        if pval.pvalue < 0.05/15: #Bonferroni adjustment (there are 15 unique pairs of feeds
            print(key, key2, pval.pvalue)
    counter += 1
# print(pvalmx)
```
