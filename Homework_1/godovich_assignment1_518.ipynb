```python
#a
cells = 250000 #(cells)
moi = 1.5 #(dimensionless)
titer = 2e8 #(infectious units per mL)
inoc_vol_ul = 500 #(Î¼L total inoculation volume)
```

```python
#b
iu_needed = cells*moi
print(iu_needed)
```

MOI (multiplicity of infection) is the number of infectious units per cell 

```python
#c
titer_ul = titer / 1000
print('Titer iu per microliter is', titer_ul)
virus_vol_needed = iu_needed / titer_ul
print('Volume of titer needed:',virus_vol_needed, 'ul')
```

```python
#d
media_vol = inoc_vol_ul - virus_vol_needed
print('Additional media volume needed:',media_vol,'ul')
```

```python
#e
print(
    f'Cell count: {cells} \nMOI: {moi} \nIU needed: {iu_needed} \nVirus titer volume: {virus_vol_needed} \nAdditional Media Volume:{media_vol}'
)
```
